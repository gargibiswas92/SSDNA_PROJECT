```python
import linecache 

file = 'repA_dna235.pdb'

outfile = open('repA_dna97.pdb', 'w')

for i in range(1, 5426):
    line = linecache.getline(file, i)
    outfile.writelines(line)
 
count = 5426   
for j in range(5426, 9933):
    line = linecache.getline(file, j)
    init = line[0:4]
    atnum = int(line[4:11])
    mid = line[11:22]
    resnum = int(line[22:26])
    end = line[26:-1]

    if resnum >= 57 and resnum <=153:
        new_res = resnum -56
        ss = '{}{:>7d}{}{:>4d}{}{}'.format(init, count, mid, new_res,end, '\n')
        outfile.writelines(ss)
        count = count + 1
    
outfile.close()
```
