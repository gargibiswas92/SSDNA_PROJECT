```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def scaling_law(x, A, v):
    return np.log(A)+v*x


data_dna = np.loadtxt('dna_med_val_0.txt')
data_rpa = np.loadtxt('rpa_dna_A_med_val_0.txt')
data_dna_c = np.loadtxt('dna_c_med_val_0.txt')
data_dna_chst = np.loadtxt('dna_c_hst_med_val_0.txt')
data_dna_cst = np.loadtxt('dna_c_st_med_val_0.txt')

fig, axp = plt.subplots(figsize=(25, 10))

length_m = [25, 30, 40, 45, 60, 80, 100]
length = [15, 25, 30, 40, 45, 60, 80, 100]
length_mm = [15, 25, 30, 40, 45, 60, 80]

axp.scatter(np.log(length), np.log(data_dna), label='ssDNA',
            color='black', s=200, edgecolor='black')
axp.scatter(np.log(length_m), np.log(data_dna_c), label='ssDNA i-i+23 --> 55 \u212B',
            color='crimson', s=200, edgecolor='black')
axp.scatter(np.log(length_m), np.log(data_dna_chst), label='ssDNA i-i+11 --> 66 \u212B',
            color='cornflowerblue', s=200, edgecolor='black')
axp.scatter(np.log(length_mm), np.log(data_rpa), label='RPA+ssDNA',
            color='gray', s=200, edgecolor='black')
axp.scatter(np.log(length_m), np.log(data_dna_cst), label='ssDNA i-i+23 --> 122 \u212B', color='mediumseagreen', s=200, edgecolor='black')
#axp.legend(fontsize=24, loc='upper left')

axp.set_title('dihedral = 0.0', fontsize=30, fontweight='bold')
axp.set_ylabel('log(R) [R= End-to-end distance in nm]', fontsize=28)
axp.set_xlabel('log(nt) [nt = number of nucleotides]', fontsize=28)
axp.xaxis.tick_bottom()
axp.yaxis.tick_left()
axp.tick_params(axis='both', which='major', labelsize=22)

spines_to_keep = ['bottom', 'left']
spines_to_remove = ['top', 'right']

for spine in spines_to_remove:
    axp.spines[spine].set_visible(False)

   
params_d, covar_d = curve_fit(scaling_law, np.log(length), np.log(data_dna))
params_cd, covar_cd = curve_fit(scaling_law, np.log(length_m), np.log(data_dna_c))
params_chstd, covar_chstd = curve_fit(scaling_law, np.log(length_m), np.log(data_dna_chst))
params_r, covar_r = curve_fit(scaling_law, np.log(length_mm), np.log(data_rpa))
params_cstd, covar_cstd = curve_fit(scaling_law, np.log(length_m), np.log(data_dna_cst))

A_d, v_d = params_d
A_cd, v_cd = params_cd
A_chstd, v_chstd = params_chstd
A_r, v_r = params_r
A_cstd, v_cstd = params_cstd

y_fit_d = scaling_law(np.log(length), A_d, v_d)
y_fit_cd = scaling_law(np.log(length_m), A_cd, v_cd)
y_fit_chstd = scaling_law(np.log(length_m), A_chstd, v_chstd)
y_fit_r = scaling_law(np.log(length_mm), A_r, v_r)
y_fit_cstd = scaling_law(np.log(length_m), A_cstd, v_cstd)

axp.plot(np.log(length), y_fit_d, color='black', label='$\\nu$ = {:>3.2f}, R0 ={:>3.2f}'.format(v_d, A_d), linewidth=6)
axp.plot(np.log(length_m), y_fit_cd, color='crimson', label='$\\nu$ = {:>3.2f}, R0={:>3.2f}'.format(v_cd, A_cd), linewidth=6)
axp.plot(np.log(length_m), y_fit_chstd, color='cornflowerblue', label='$\\nu$ = {:>3.2f}, R0={:>3.2f}'.format(v_chstd, A_chstd), linewidth=6)
axp.plot(np.log(length_mm), y_fit_r, color='gray',label='$\\nu$ = {:>3.2f}, R0={:>3.2f}'.format(v_r, A_r), linewidth=6)
axp.plot(np.log(length_m), y_fit_cstd, color='mediumseagreen', label='$\\nu$ = {:>3.2f}, R0={:>3.2f}'.format(v_cstd, A_cstd), linewidth=6)
axp.legend(fontsize=20, loc='lower right', ncol=2)

fig.savefig('all_models_log_scale_0.png', dpi=300, bbox_inches='tight')
```
