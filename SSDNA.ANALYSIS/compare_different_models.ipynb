```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 


list = ['40_0_HIGHSALT', '40_0_LOWSALT', '40_7_HIGHSALT', '40_7_LOWSALT']
length = ['0.0 (salt = 0.1)', '0.0 (salt = 0.01)' , '0.7 (salt = 0.1)', '0.7 (salt = 0.01)']

xtick_pos = np.array([0, 1, 2, 3])
cmap = sns.color_palette("Set2")

data0, median_dna, std_dna =[], [], []

for ll in list:
    name="/trajectories/gargi/RPA_Ed/RESULTS.ANALYSIS/end_to_end_"+ll+".txt"
    data = np.loadtxt(name)
    mask = np.isnan(data)
    data_n = data[~mask]
    #data_n = data_n[data_n < 100.0]
    dt=data_n/10
    me1=np.median(dt)
    sd1=np.std(dt)
    median_dna.append(me1)
    std_dna.append(sd1)
    data0.append(dt)
    
fig, ax = plt.subplots(figsize=(12,8))
    
sns.violinplot(data=data0, width=0.8, ax=ax, palette=cmap, split=True, scale="width")
ax.set_xticks(xtick_pos, length, fontsize=24, rotation = 45)
ax.set_ylabel('Distance (nm)', fontsize=28)

ax.set_title('ssDNA', fontsize=30, fontweight='bold')
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.tick_params(axis='both', which='major', labelsize=22)

spines_to_keep = ['bottom', 'left']
spines_to_remove = ['top', 'right']

for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
fig.savefig('compare_salt_dihedral.png', dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 


list = ['40_0', "40_0_HIGHSALT" ,"40_0_NEU_PHOS", "40_0_NEU_PHOS_NO_DIH", "40_0_NEU_PHOS_NO_DIH_CH_REP"]
length = ['dih = 0.0', 'dih = 0.0\nhighsalt', 'dih = 0.0\nphos_ch =0', 'all dih = 0.0\nphos_ch =0', 'all dih = 0.0\nphos_ch =0\nsmaller bead' ]

xtick_pos = np.array([0, 1, 2, 3, 4])
cmap = sns.color_palette("Set2")

data0, median_dna, std_dna =[], [], []

for ll in list:
    name="/trajectories/gargi/RPA_Ed/RESULTS.ANALYSIS/end_to_end_"+ll+".txt"
    data = np.loadtxt(name)
    mask = np.isnan(data)
    data_n = data[~mask]
    #data_n = data_n[data_n < 100.0]
    dt=data_n/10
    me1=np.median(dt)
    sd1=np.std(dt)
    median_dna.append(me1)
    std_dna.append(sd1)
    data0.append(dt)
    
fig, ax = plt.subplots(figsize=(12,8))
    
sns.violinplot(data=data0, width=0.8, ax=ax, palette=cmap, split=True, scale="width")
ax.set_xticks(xtick_pos, length, fontsize=24, rotation=45)
ax.set_ylabel('Distance (nm)', fontsize=28)

ax.set_title('ssDNA', fontsize=30, fontweight='bold')
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.tick_params(axis='both', which='major', labelsize=22)

spines_to_keep = ['bottom', 'left']
spines_to_remove = ['top', 'right']

for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
fig.savefig('compare_diff_models_40_nt.png', dpi=300, bbox_inches='tight')
```
