```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


list = ['side', 'face']
length = [60, 80, 100]

xtick_pos = np.array([0, 1, 2])
cmap = sns.color_palette("Set2")

data_4s, median_dna_4s, std_dna_4s = [], [], []
data_4f, median_dna_4f, std_dna_4f = [], [], []

for ll in length:
    name_4s = "/trajectories/gargi/RPA_Ed/RESULTS.ANALYSIS/end_to_end_TWO_RPA_stacks_side_" + \
        str(ll)+"_dih_0.4.txt"
    name_4f = "/trajectories/gargi/RPA_Ed/RESULTS.ANALYSIS/end_to_end_TWO_RPA_stacks_face_" + \
        str(ll)+"_dih_0.4.txt"
    data_4sn = np.loadtxt(name_4s)
    data_4fn = np.loadtxt(name_4f)
    mask = np.isnan(data_4sn)
    data_n4s = data_4sn[~mask]
    maskp = np.isnan(data_4fn)
    data_n4f = data_4fn[~maskp]
    dt_4s = data_n4s/10
    dt_4f = data_n4f/10
    me4s, me4f = np.mean(dt_4s), np.mean(dt_4f)
    sd4s, sd4f = np.std(dt_4s), np.mean(dt_4f)
    median_dna_4s.append(me4s)
    std_dna_4s.append(sd4s)
    data_4s.append(dt_4s)
    median_dna_4f.append(me4f)
    std_dna_4f.append(sd4f)
    data_4f.append(dt_4f)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

sns.violinplot(data=data_4s, width=0.8,
               ax=ax[0], palette=cmap, split=True, scale="width")
sns.violinplot(data=data_4f, width=0.8,
               ax=ax[1], palette=cmap, split=True, scale="width")

ax[0].set_title('side-by-side', fontsize=30, fontweight='bold')
ax[1].set_title('face-to-face', fontsize=30, fontweight='bold')
for axp in ax:
    axp.set_xticks(xtick_pos, length, fontsize=24,
                   fontweight='bold', rotation=45)
    axp.set_ylabel('Distance (nm)', fontsize=28)
    axp.xaxis.tick_bottom()
    axp.yaxis.tick_left()
    axp.tick_params(axis='both', which='major', labelsize=22)

    spines_to_keep = ['bottom', 'left']
    spines_to_remove = ['top', 'right']

    for spine in spines_to_remove:
        axp.spines[spine].set_visible(False)
    axp.set_ylim(0,27)
#fig.savefig('length_vs_distance_rpa_dna_violin1.png', dpi=300, bbox_inches='tight')
```
