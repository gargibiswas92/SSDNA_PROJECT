```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 


list = [0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0.0]
length = ['k$_{D}$ = 0.6','k$_{D}$ = 0.5', 'k$_{D}$ = 0.4', 'k$_{D}$ = 0.3', 'k$_{D}$ = 0.2', 'k$_{D}$ = 0.1', 'k$_{D}$ = 0.0']

xtick_pos = np.array([0, 1, 2, 3, 4, 5, 6])
cmap = sns.color_palette("Set2")

data0_16, median_dna_16, std_dna_16 =[], [], []

for ll in list:
    name="/trajectories/gargi/RPA_Ed/RESULTS.ANALYSIS/end_to_end_rad_1.6_dih_"+str(ll)+".txt"
    data = np.loadtxt(name)
    mask = np.isnan(data)
    data_n = data[~mask]
    #data_n = data_n[data_n < 100.0]
    dt=data_n/10
    me1=np.mean(dt)
    sd1=np.std(dt)
    median_dna_16.append(me1)
    std_dna_16.append(sd1)
    data0_16.append(dt)
    
fig, ax = plt.subplots(figsize=(12,8))
    
sns.violinplot(data=data0_16, width=0.8, ax=ax, palette=cmap, split=True, scale="width")
ax.set_xticks(xtick_pos, length, fontsize=24, fontweight='bold', rotation=45)
ax.set_ylabel('Distance (nm)', fontsize=28)

ax.set_title('Radius = 1.6', fontsize=30, fontweight='bold')
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.tick_params(axis='both', which='major', labelsize=22)

spines_to_keep = ['bottom', 'left']
spines_to_remove = ['top', 'right']

for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
fig.savefig('dihedral_end_to_end_violin.png', dpi=300, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(12,8))
    
x = [0, 1, 2,3, 4, 5, 6]
ax.bar(x, median_dna_16, width=0.25, yerr=std_dna_16, capsize=6)
ax.set_xticks(xtick_pos, length, fontsize=24, fontweight='bold', rotation =45)
ax.set_ylabel('Distance (nm)', fontsize=28)

ax.set_title('Radius = 1.6', fontsize=30, fontweight='bold')
ax.xaxis.tick_bottom()
ax.yaxis.tick_left()
ax.tick_params(axis='both', which='major', labelsize=22)

spines_to_keep = ['bottom', 'left']
spines_to_remove = ['top', 'right']

for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
ax.set_ylim(-1, 10)
fig.savefig('dihedral_end_to_end_bar.png', dpi=300, bbox_inches='tight')
```
